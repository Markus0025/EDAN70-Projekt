```python
import json
from tqdm import tqdm
```

```python
with open('2_TO_WIKI.json', encoding = "utf-8") as f:
    wiki_2 = json.load(f)
    f.close()

with open('3_TO_WIKI.json', encoding = "utf-8") as f:
    wiki_3 = json.load(f)
    f.close()
```

```python
print(len(wiki_2))
print(len(wiki_3))
print(wiki_2[2])
```

```python
wiki_2_cut = []
thresh = 0.9
for e in wiki_2:
    if e["similarity_score"] > thresh:
        wiki_2_cut.append(e)

print(len(wiki_2_cut))
print(len(wiki_2))
```

```python
wiki_3_cut = []
thresh = 0.9
for e in wiki_3:
    if e["similarity_score"] > thresh:
        wiki_3_cut.append(e)

print(len(wiki_3_cut))
print(len(wiki_3))
```

```python
with open('different_QIDs.json', encoding = "utf-8") as f:
    lookup = json.load(f)
    f.close()
```

```python
print(len(lookup))
```

```python
print(lookup[0])
lookup_dict = {l[1]: l[0] for l in lookup}
```

```python
wiki_2_linked = []
men_2 = 0
women_2 = 0

for e in tqdm(wiki_2_cut):
    if e["match_gender"] == "Man":
        men_2 += 1
    elif e["match_gender"] == "Woman":
        women_2 += 1

    ent = e.copy()
    qid = lookup_dict.get(e["match_headword"])

    if qid:
        ent["match_qid"] = qid
        ent["match_link"] = f"https://www.wikidata.org/wiki/{qid}"
    else:
        ent["match_qid"] = ""
        ent["match_link"] = ""

    wiki_2_linked.append(ent)


print(len(wiki_2_linked))
print(len(wiki_2_cut))
print("Number of men: ", men_2)
print("Number of women: ", women_2)
print("Ed 2: men: ",men_2/(men_2+women_2), ", women: ", women_2/(men_2+women_2))

wiki_3_linked = []
men_3 = 0
women_3 = 0

for e in tqdm(wiki_3_cut):
    if e["match_gender"] == "Man":
        men_3 += 1
    elif e["match_gender"] == "Woman":
        women_3 += 1

    ent = e.copy()
    qid = lookup_dict.get(e["match_headword"])

    if qid:
        ent["match_qid"] = qid
        ent["match_link"] = f"https://www.wikidata.org/wiki/{qid}"
    else:
        ent["match_qid"] = ""
        ent["match_link"] = ""

    wiki_3_linked.append(ent)


print(len(wiki_3_linked))
print(len(wiki_3_cut))
print("Number of men: ", men_3)
print("Number of women: ", women_3)
print("Ed 3: men: ",men_3/(men_3+women_3), ", women: ", women_3/(men_3+women_3))
```

```python
with open('ED_2_LINKED.json', 'w', encoding='utf-8') as f:
    json.dump(wiki_2_linked, f, ensure_ascii=False, indent=4)
    f.close()

with open('ED_3_LINKED.json', 'w', encoding='utf-8') as f:
    json.dump(wiki_3_linked, f, ensure_ascii=False, indent=4)
    f.close()
```
