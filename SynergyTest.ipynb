```python
import json
from tqdm import tqdm
```

```python
with open('3_TO_2_KB.json', encoding='utf8') as f:
    kb = json.load(f)
    f.close()
```

```python
with open('2TO3_MATCHES_NEW_ENCODER.json', encoding='utf8') as f:
    a = json.load(f)
    f.close()

with open('3TO2_MATCHES_NEW_ENCODER.json', encoding='utf8') as f:
    t = json.load(f)
    f.close()
```

```python
kb_thresh_val = []
thresh = 0.9
for ent in kb:
    if ent["similarity_score"] > thresh:
        kb_thresh_val.append(ent)
    else: 
        ent2 = ent.copy()
        ent2["match_id"] = "nothing"
        kb_thresh_val.append(ent2)

print(len(kb_thresh_val))
print(len(kb))
```

```python
andra_thresh = []
andra_thresh_val = []
thresh = 0.925
for ent in a:
    if ent["similarity_score"] > thresh:
        andra_thresh.append(ent)
        andra_thresh_val.append(ent)
    else: 
        ent2 = ent.copy()
        ent2["match_id"] = "nothing"
        andra_thresh_val.append(ent2)

tredje_thresh = []
tredje_thresh_val = []
thresh = 0.925
for ent in t:
    if ent["similarity_score"] > thresh:
        tredje_thresh.append(ent)
        tredje_thresh_val.append(ent)
    else:
        ent2 = ent.copy()
        ent2["match_id"] = "nothing"
        tredje_thresh_val.append(ent2)
```

```python
print(len(andra_thresh))
print(len(andra_thresh_val))
print(len(tredje_thresh))
print(len(tredje_thresh_val))
```

```python
synergy_matches = []

for m23 in tqdm(andra_thresh):
    ent_id = m23["entity_id"]
    match_id = m23["match_id"]
    for m32 in tredje_thresh:
        if match_id == m32["entity_id"] and ent_id == m32["match_id"]:
            synergy_matches.append(m23)
            break

print(len(synergy_matches))
```

```python
with open('SYNERGY_MATCHES_NEW_ENCODER.json', 'w', encoding='utf-8') as f:
    json.dump(synergy_matches, f, ensure_ascii=False, indent=4)
    f.close()
```

```python
with open('PERSON_MATCHING_VAL_DATA.json', encoding='utf8') as f:
    val = json.load(f)
    f.close()
```

```python
validation = []
test = []
i = 0
for e in val:
    if i % 2 == 0:
        validation.append(e)
    else:
        test.append(e)
    i += 1

print(len(validation))
print(len(test))
```

```python
tp = 0
fp = 0
tn = 0
fn = 0


for m in tredje_thresh_val:
    qid = m["entity_id"]
    for val_matches in test:
        if qid == val_matches["id1"]:
            if m["match_id"] == val_matches["id2"] and m["match_id"] != "nothing":
                tp += 1
            elif m["match_id"] == val_matches["id2"] and m["match_id"] == "nothing":
                tn += 1
            else:
                if m["match_id"] == "nothing":
                    fn += 1
                else:
                    fp += 1



precision = tp/(tp+fp)
recall = tp/(tp+fn)
f1 = (2*precision*recall)/(precision+recall)
print("Precision: ", precision, " Recall: ", recall, " F1-score: ", f1)
```
